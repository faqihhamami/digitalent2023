```python
# operasi aritmatika 3 pilihan

pilihan = input("masukkan pilhan anda? (A atau B atau C)")
a = 5
b = 3

if pilihan == "A":
  print("hasilnya", a+b)
elif pilihan == "B":
  print("hasilnya", a-b)
elif pilihan == "C":
  print("hasilnya", a*b)
else:
  print("input anda salah")
```

```python
while True:
  pilihan = input("masukkan pilhan anda? (A atau B atau C)")
  a = 5
  b = 3

  if pilihan == "A":
    print("hasilnya", a+b)
  elif pilihan == "B":
    print("hasilnya", a-b)
  elif pilihan == "C":
    print("hasilnya", a*b)
  else:
    print("input anda salah. program berheti")
    break
```
