## Function with Parameter

```python
def cetak(nama):
  print("selamat pagi", nama)
  print("bagaimana kabar anda", nama)
```

```python
cetak("Tono")
```

```python
cetak("Rani")
```

```python
def salam(nama, umur):
  print("selamat pagi", nama)
  print("umur anda adalah",umur)
  print(f"tipe nama {type(nama)} dan tipe umur {type(umur)}")
```

```python
salam("tono", 18)
```

- misal A = [1,2,3,4,5]

```python
a = [1,2,3,4,5]

rata2 = sum(a)/len(a)
print(rata2)
```

## Return Value

```python
# operasi penjumlahan dari 2 rata2 list
```

```python
a = [2,3,4,6,7]
b = [2,5,1,4,6]
```

```python
def rata2(x):
  rata = sum(x) / len(x)
  return rata
```

```python
rataA = rata2(a)
rataB = rata2(b)
print(f"hasil pemjumlahan dari 2 rata2 {a} dan {b} adalah {rataA+rataB}")
```
