## Map function

```python
kata = 'aono'

def ambilFirst(s):
  return s[0] == 'a'
```

```python
ambilFirst(kata)
```

```python
nama = ['tono', 'andi', 'zaki', 'anton', 'ana']

mapobject = map(ambilFirst, nama)
list(mapobject)
```

```python
# menggunakan lambda
list(map(lambda s:s[0] == 'a', nama))
```

## Filter Function

```python
filterobject = filter(ambilFirst, nama)
list(filterobject)
```

```python
# menggunakan lambda
list(filter(lambda s:s[0] == 't', nama))
```

## Lab 3

- Tampilkan item numeric dari list berikut = `["3","1","Toni","5"]`
- Gunakan fungsi `map & filter`
- Gunakan Lambda expression

```python
listA = ["3","1","Toni","5"]

cek = map(lambda a : a.isnumeric(), listA)
list(cek)
```

```python
cek = filter(lambda a : a.isnumeric(), listA)
list(cek)
```

## File Operation

```python
# ada 3 mode : write, read, append
```

```python
with open('satu.txt', 'w') as file:
  file.write("ini adalah kalimat satu")
```

```python
# read
with open('satu.txt', 'r') as file:
  hello = file.read()
  print(hello)
```

```python
# append
with open('satu.txt', 'a') as file:
  file.write("\nini adalah kalimat dua")
```
