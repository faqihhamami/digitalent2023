## Try... Except... Else

```python
# exception untuk pembagian 0
try:
  a = int(input("masukkan a : "))
  b = int(input("masukkan b : "))

  c = a/b
  print(f'hasilnya adalah {c}')
except Exception as e:
  print("anda salah masukkan input. errornya adalah {}".format(e))
else:
  print("baris ini dicetak")
```

## Try... Except... Else... Finally

```python
# exception untuk pembagian 0
try:
  a = int(input("masukkan a : "))
  b = int(input("masukkan b : "))

  c = a/b
  print(f'hasilnya adalah {c}')
except Exception as e:
  print("anda salah masukkan input. errornya adalah {}".format(e))
else:
  print("baris ini dicetak")
finally:
  print("baris ini akan selalu dicetak di kondisi apapun")
```

## Lab 2

-  ada list of string yang berisi `['18/07/2023 18:19:30', '19/07/2023 20:29:30', '20/07/2023 10:12:10']`
- Simpan masing2 `tahun, bulan, tanggal dan jam` di list yang berbeda. Misalnya `listTanggal = ['18','19','20']`

```python
tanggal = '18/07/2023 18:19:30'

d = tanggal.split()[0].split('/')[0]
m = tanggal.split()[0].split('/')[1]
y = tanggal.split()[0].split('/')[2]
h = tanggal.split()[1].split(':')[0]

print(y,m,d,h)
```

```python
listdate = ['18/07/2023 18:19:30', '19/07/2023 20:29:30', '20/07/2023 10:12:10']

listy = []
listm = []
listd = []
listh = []

for i in listdate:
  d = i.split()[0].split('/')[0]
  m = i.split()[0].split('/')[1]
  y = i.split()[0].split('/')[2]
  h = i.split()[1].split(':')[0]

  # print(y,m,d,h)
  listy.append(y)
  listm.append(m)
  listd.append(d)
  listh.append(h)

print(listy)
print(listm)
print(listd)
print(listh)
```

```python
# menggunakan list comprehension
listdate = ['18/07/2023 18:19:30', '19/07/2023 20:29:30', '20/07/2023 10:12:10']


listd = [i.split()[0].split('/')[0] for i in listdate]
listm = [i.split()[0].split('/')[1] for i in listdate]
listy = [i.split()[0].split('/')[2] for i in listdate]
listh = [i.split()[1].split(':')[0] for i in listdate]

print(listy)
print(listm)
print(listd)
print(listh)
```
