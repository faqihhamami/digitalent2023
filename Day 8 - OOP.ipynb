```python
class Mahasiswa:
  def __init__(self, nim, nama, jurusan):
    self.__nim = nim
    self.__nama = nama
    self.__jurusan = jurusan
    self.__nilai = {}

  def tampilInfo(self):
    print(f'nama {self.__nama}, nim {self.__nim}, jurusan {self.__jurusan}')

  # nilai['alpro'] = 60
  # nilai['kalkulus'] = 90
  def tambah_nilai(self, matakuliah, nilaimk):
    self.__nilai[matakuliah] = nilaimk

  def get_nilai(self):
    print(self.__nilai)

  def get_rata(self):
    rata2 = sum(self.__nilai.values()) / len(self.__nilai)
    return rata2
```

```python
mahasiswa1 = Mahasiswa('12346', 'Rani', 'informatika')

mahasiswa1.tampilInfo()
```

```python
mahasiswa1.tambah_nilai('kalkulus', 90)
```

```python
mahasiswa1.get_nilai()
```

```python
mahasiswa1.tambah_nilai('alpro', 88)
mahasiswa1.tambah_nilai('matdis', 80)

mahasiswa1.get_nilai()
```

```python
mahasiswa2 = Mahasiswa('12348', 'Diki', 'informatika')
mahasiswa2.tambah_nilai('kalkulus', 66)
mahasiswa2.tambah_nilai('alpro', 77)
mahasiswa2.tambah_nilai('matdis', 88)

mahasiswa2.tampilInfo()
```

```python
print(mahasiswa1.get_nilai())
print(mahasiswa2.get_nilai())
```

## Inheritance

```python
class Unggas:
  def __init__(self):
    print('hewan unggas')

  def berenang(self):
    print("swim faster")
```

```python
class Pinguin(Unggas):
  def __init__(self):
    super().__init__()
    print("pinguin siap")

  def berenang(self):
    print("berenang sangat cepat")
```

```python
unggas1 = Unggas()
unggas1.berenang()
```

```python
unggas2 = Pinguin()
```

```python
unggas2.berenang()
```

## Case OOP di Penerbitan

```python
class Penulis:
  def __init__(self, nama, alamat):
    self.nama = nama
    self.alamat = alamat

  def identitas(self):
    print(f'nama {self.nama} alamat {self.alamat}')
```

```python
class Buku:
  def __init__(self, penulis, judul, kategori):
    self.penulis = penulis
    self.judul = judul
    self.kategori = kategori
```

```python
penulis1 = Penulis('Dani', 'Bandung')
```

```python
penulis1.identitas()
```

```python
penulis2 = Penulis('Ana', 'Jakarta')
penulis2.identitas()
```

```python
buku1 = Buku(penulis1, 'belajar python', 'tutorial')
```

```python
buku1.penulis.identitas()
```
