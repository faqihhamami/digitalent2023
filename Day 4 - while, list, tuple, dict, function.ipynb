## List & Tuple

```python
# list -> mutable
mahasiswa = ['Tono', 18, 158.8, True]
```

```python
mahasiswa[0]
```

```python
mahasiswa[1] = 20
mahasiswa[1]
```

```python
# Fungsi di List
# 1. append -> menambah item ke dalam list

mahasiswa.append('bandung')
mahasiswa
```

```python
# 2. len -> mengetahui jumlah item

len(mahasiswa)
```

```python
# 3. pop -> menghapus item terakhir (LIFO)

mahasiswa.pop()
mahasiswa
```

```python
# tuple -> immutable
student = ('andi', 19, 166.6, True)
```

```python
student[0]
```

```python
student[1] = 21
student[1]
```

```python
len(student)
```
